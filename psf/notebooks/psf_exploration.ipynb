# NOTES

This notebooks explores WebbPSF. It calculates monochromatic PSF at 100 wavelengths evenly spaced from 10000 to 20000 Angstroms. At the end, it calculates a PSF using the "GRISM0" filter. 

JWST classes (i.e. NIRSpec) can run calc_datacube_fast to do our monochromatic for-loop quickly. This does not exist for the roman WFI class. It assumes the "exit pupil wavefront is independent of wavelength." The claim is this is valid for all but coronographic and slit spectroscopic PSFs.

```python
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = (16, 7)  # make the default figure size larger
matplotlib.rcParams['image.interpolation'] = 'nearest'  # don't blur/smooth image plots
from matplotlib import pyplot as plt
from matplotlib.colors import LogNorm
from matplotlib import animation
import webbpsf
import webbpsf.roman

from scipy import constants
import numpy as np
from tqdm import tqdm

webbpsf.setup_logging()
```

```python
wfi = webbpsf.roman.WFI()
wfi.filter = "GRISM0"
```

```python
psf_0_0[0].data.sum()

# Maybe functional amplitdue functions?
f158_lam_center = 1.577 * 10**-6 # micrometer
telarea = 2.36**2 * constants.pi # meter ** 2

f_0 = 3.7059108273831045e-10 # this is just a number that seems to work; theoretically, this is comparable to an ABZeropoint

# Calculate number of counts as a function of magnitude
N = lambda mag: telarea * (f158_lam_center / (constants.h * constants.speed_of_light)) * f_0 * 10**(-0.4 * mag)

N(13.5)
```

```python
psf_0_0[0].data.max()
```

```python
# Exploring Magnitude adjustments and Spectral Source File
src = webbpsf.specFromSpectralType('G0V', catalog='phoenix')

wfi.detector_position = (4, 4)
psf_0_0 = wfi.calc_psf(normalize='last', fov_pixels=364, oversample=2, source=src)

factor = N(13.5)

vmin = 1e-7 * factor
vmax = 0.1 * factor
psf_0_0[0].data *= factor
plt.imshow(psf_0_0[0].data, interpolation='nearest', origin='lower', norm=LogNorm(vmin=vmin, vmax=vmax), cmap='hot')
psf_0_0[0].data.max()
```

```python
# Comapre PSF at opposite ends of SCA1
src = webbpsf.specFromSpectralType('G0V', catalog='phoenix')

wfi.detector_position = (4, 4)
psf_0_0 = wfi.calc_psf(normalize='last', fov_pixels=364, oversample=2, source=src)

wfi.detector_position = (4092,4092)
psf_4088_4088 = wfi.calc_psf(normalize='last', fov_pixels=364, oversample=2, source=src)


fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
webbpsf.display_psf(psf_0_0, ax=ax1, title="PSF_0_0")
webbpsf.display_psf(psf_4088_4088, ax=ax2, title="PSF_4088_4088")
webbpsf.display_psf_difference(psf_0_0, psf_4088_4088, ax=ax3, title="diff")
```

```python
# Calculate PSFs at all wavelengths
webbpsf.setup_logging(level="ERROR")
src = webbpsf.specFromSpectralType('G0V', catalog='phoenix')
wave_range = np.arange(10000, 20000, 100)

wave_dependent = {}

for ii in tqdm(wave_range):
    key = str(ii)
    wave_dependent[key] = wfi.calc_psf(monochromatic=(ii*(10**-10)), fov_pixels=364, oversample=2, source=src)
```

```python
# Animate how PSF changes with wavelength
%matplotlib widget

fig, ax = plt.subplots()

ims = []
for key in wave_dependent.keys():
    im = ax.imshow(wave_dependent[key][0].data, animated=True, interpolation='nearest', origin='lower', norm=LogNorm(vmin=1e-7, vmax=0.1), cmap='hot')
    if ii == "10000":
        ax.imshow(wave_dependent[key][0].data, interpolation='nearest', origin='lower', norm=LogNorm(vmin=1e-7, vmax=0.1), cmap='hot')
    ims.append([im])


ani = animation.ArtistAnimation(fig, ims, interval=50, blit=True,
                                repeat_delay=1000)

plt.show()
```
